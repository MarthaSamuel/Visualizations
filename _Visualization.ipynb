```python
#B). Next step - create a figure

# figure is the top level container where everything in the plot is drawn
# facecolor - background color
# setting the size of the figure - figaspect(height:breadth ratio)#we can say;figsize=plt.figaspect(2)

fig=plt.figure(facecolor='yellow',figsize=(4,4))

#C). then create axis for the figure.fig.add_subplot(nrows,ncols,no of fig)
ax=fig.add_subplot(1,1,1)

# for experiment

#ax=fig.add_subplot(1,2,1)
#ax=fig.add_subplot(1,1,2)
#ax=fig.add_subplot(2,1,1)
#ax=fig.add_subplot(1,2,2)
#ax=fig.add_subplot(2,2,2)
#ax=fig.add_subplot(2,2,1)

#D). we add arguments to the axis,set axis object parameter -xlimit, ylimit,xlabel, ylabel
'''we can say
#explicit passing of args
ax.set_xlim([1,5])
ax.set_ylim([3,10])
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_title('my plot')
'''
#this is implicit
ax.set(xlim=[1,5],ylim=[3,10],xlabel='x-axis',ylabel='y- axis',title='my first plot')

#E). plot the points

a=[2,3,4]
b=[6,8,1]

#F). now plot a on the axis and b on y axis by a straight line
# plot function will join the points by a straight line
plt.plot(a,b)
```

## Multiple plots

```python
# MULTIPLE PLOTS

fig=plt.figure(facecolor='yellow',figsize=(7,7))

# 1 row , 3 columns, 3 figures

# set the 1st figure
fig1=fig.add_subplot(1,3,1)
fig1.plot(a,b)
fig1.set(xlabel='x axis',ylabel='y axis')

fig2=fig.add_subplot(1,3,2)
fig2.scatter(a,b)

fig3=fig.add_subplot(1,3,3)
fig3.bar(a,b)
```

## Subplots

```python
# SUBPLOT

# subplots - fig and an axes object

# 2 by 2 matrix - 2 x axis and 2 y axis - total plots 4

# (figure,axes) =plt.add_subplots(no of rows, no of columns)

fig,axes=plt.subplots(facecolor='orange',nrows=2,ncols=2,sharex=False,sharey=False)

axes[0,0].set(title='1st graph - left upper')
axes[0,0].plot([1,2,3],[10,20,30])

# axes object - numpy array 
axes[0,1].set(title='2nd graph - right upper')
axes[0,1].scatter(a,b)

axes[1,0].set(title='3rd graph - left lower')

axes[1,1].set(title='4th graph - right lower')

plt.tight_layout() #this makes fig less tight
```
